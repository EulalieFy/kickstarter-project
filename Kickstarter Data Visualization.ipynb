```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/Users/eulalieformery/Desktop/Datacamp/kickstarter-projects/ks-projects-201801.csv")
data = data.dropna(subset=['name'])
```

```python
data['Target'] = np.where( data['usd_pledged_real']>=data['usd_goal_real'], 1, 0)
```

```python
data['launched_date'] = pd.to_datetime(data['launched'], format='%Y-%m-%d %H:%M:%S')
data['deadline_date'] = pd.to_datetime(data['deadline'], format='%Y-%m-%d %H:%M:%S')
```

```python
data['year'] = [d.year for d in data['launched_date']]
data['month'] = [d.month for d in data['launched_date']]
```

# State of projects

```python
groupby_state = pd.DataFrame(data.groupby('state').count()['ID'])
groupby_state = groupby_state.sort_values(by=['ID'], ascending = False)
```

```python
x = np.arange(6)
state = groupby_state.ID.tolist()

fig, ax = plt.subplots()

plt.bar(x, state)
plt.xticks(x, ('failed', 'successful', 'canceled', 'undefined', 'live', 'suspended'))
plt.show()
```

# Categories

Most represented categories

```python
groupby_cat = pd.DataFrame(data.groupby('category').count()['ID'])
groupby_cat = groupby_cat.sort_values(by=['ID'], ascending = False)
```

```python
groupby_cat_15 = groupby_cat.head(15)
```

```python
x = np.arange(15)
state = sorted(groupby_cat_15.ID.tolist(), reverse= True)

fig, ax = plt.subplots( figsize= (20,10))

plt.bar(x, state)
plt.xticks(x, ('Product Design',
 'Documentary',
 'Music',
 'Tabletop Games',
 'Shorts',
 'Video Games',
 'Food',
 'Film & Video',
 'Fiction',
 'Fashion',
 'Nonfiction',
 'Art',
 'Apparel',
 'Theater',
 'Technology'))
plt.show()
```
