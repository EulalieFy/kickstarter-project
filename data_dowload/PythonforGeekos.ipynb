```python
import boto3
import os
import progressbar

print(os.getcwd())

link = 'https://s3.amazonaws.com/weruns/forfun/Kickstarter/Kickstarter_2018-12-13T03_20_05_701Z.zip'

s3 = boto3.resource('s3')
bucket = s3.Bucket('Kickstarter_2018-12-13T03_20_05_701Z.zip')

from botocore.handlers import disable_signing
resource = boto3.resource('s3')
resource.meta.client.meta.events.register('choose-signer.s3.*', disable_signing)

BUCKET_NAME = 'weruns' # replace with your bucket name
KEY = 'forfun/Kickstarter/Kickstarter_2018-12-13T03_20_05_701Z.zip' # replace with your object key
bucket = resource.Bucket(BUCKET_NAME)

#s3 = boto3.resource('s3')

#try:
#    s3.Bucket(BUCKET_NAME).download_file(KEY, 'Kickstarter_data.jpg')
with open('Kickstarter_data.zip', 'wb') as data:
    file_name = os.path.join(os.getcwd(),'Kickstarter_data.zip')
    statinfo = os.stat(file_name)
    up_progress = progressbar.progressbar.ProgressBar(maxval=204864714)#statinfo.st_size)
    up_progress.start()

    def upload_progress(chunk):
        up_progress.update(up_progress.currval + chunk)
        
    bucket.download_fileobj(KEY, data,  Callback=upload_progress)
    
statinfo = os.stat(file_name)
statinfo.st_size

import zipfile
zip_ref = zipfile.ZipFile(file_name, 'r')
zip_ref.extractall('./data/Kickstarter')
zip_ref.close()

print('Process Done')
```

```python
import re

def get_picture(row,size ='big'):
    if size == 'big':
        print(row['profile'])
        return re.findall(r'(?<="default":")[^"]*', row['profile'])[0]
    if size == 'small':
        return re.findall(r'(?<="baseball_card":")[^"]*', row['profile'])[0]

get_picture(df1.loc[1,:], 'small')
```

```python
from PIL import Image
import requests
import IPython.display as ipd
from io import BytesIO

from tqdm import tqdm_notebook

for row in tqdm_notebook(df1.iterrows()):

    url = get_picture(row[1],'small')
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))

    img.save("./data/Image/{}.png".format(row[1]['id']))
```

```python
import newspaper

urls : {"web":{"project":"https://www.kickstarter.com/projects/529454320/the-story-of-pweep-from-egg-to-peacock?ref=category_newest","rewards":"https://www.kickstarter.com/projects/529454320/the-story-of-pweep-from-egg-to-peacock/rewards"}}

def get_articles(url):
    
    ''' Download un article a partir d'un URL grace au module *newspaper* '''
    
    art = newspaper.Article(url)
    art.download()
    art.parse()
    return(art.text)

def download_description(row):
    if os.path.exists('./data/Description/{}.txt'.format(row['id'])) == True:
        print('Already Downloaded')
        return

    try :
        url = re.findall(r'(?<="project":")[^"]*', row['urls'])[0]
        article = get_articles(url)
        #print(row['id'])
        with open('./data/Description/{}.txt'.format(row['id']), 'w') as f:
            f.write(article)
            f.close()
    except :
        print('Error while downloading ID:', row['id'], 'url', url)
        
from tqdm import tqdm_notebook

for row in tqdm_notebook(df1.iterrows()):
    download_description(row[1])
    
#print(get_articles('https://www.kickstarter.com/projects/529454320/the-story-of-pweep-from-egg-to-peacock?ref=category_newest'))
```
